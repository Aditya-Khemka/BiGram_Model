```python
import torch 
import torch
import numpy as np
import time
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)
torch.cuda.synchronize()
```

```python
torch_rand1 = torch.rand(100, 100, 100, 100, device=device)
torch_rand2 = torch.rand(100, 100, 100, 100, device=device)
start_time = time.time()
rand_torch = torch_rand1 @ torch_rand2
end_time = time.time()
print(f"PyTorch (GPU) time: {end_time - start_time:.8f} seconds")


np_rand1 = np.random.rand(100, 100, 100, 100).astype(np.float32)
np_rand2 = np.random.rand(100, 100, 100, 100).astype(np.float32)
start_time = time.time()
rand_np = np.matmul(np_rand1, np_rand2)
end_time = time.time()
print(f"NumPy (CPU) time: {end_time - start_time:.6f} seconds")
```

```python
# Define a probability tensor
# 10% or 0.1 => 0 ; 90% or 0.9 => 1.
probabilities = torch.tensor([0.1, 0.9])

# Draw 5 samples from the multinomial distribution
samples = torch.multinomial(probabilities, num_samples=10, replacement=True)
print(samples)
```

```python
tensor1 = torch.tensor([1, 2, 3, 4])
tensor2 = torch.tensor([5, 6, 1, 4])
out = torch.cat((tensor1, tensor2), dim=0)
out
```

```python
out = torch.tril(torch.ones(5, 5)) #set everything above the diagonal as false
out
```

```python
out = torch.triu(torch.ones(5, 5)) #set everything below the diagonal as false
out
```

```python
out = torch.zeros(5, 5).masked_fill( #5x5 zero
    torch.tril(torch.ones(5, 5)) == 0, # trail keeps elements on and below the main diagonal, sets everything above it to 0
                                       #==0 creats a T/F boolean matrix (5x5)
    float('-inf')                      #Wherever mask == True, replace value with -∞
)
out
#Prevent a token from “seeing” future tokens
```
